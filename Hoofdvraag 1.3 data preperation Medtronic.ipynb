```python
# pandas inporteren, csv file ophalen en checken of dit goed gaat
import pandas as pd
medtronic= pd.read_csv("Medtronic_US.csv", delimiter=",")
medtronic.head()
```

```python
#Data set filteren op de gegevens die ik nodig heb en alle NaN values droppen
medtronic2= medtronic[['Subject code number','Local date [yyyy-mm-dd]','Local time [hh:mm:ss]','BG Reading [mmol/l]']]
medtronic2.dropna(subset = ["BG Reading [mmol/l]"], inplace=True)
medtronic2
```

```python
#Data type van Historic Glucose veranderen naar interger en checken of dit goed gegaan 
medtronic2= medtronic2.astype({"BG Reading [mmol/l]": int})
print(medtronic2.dtypes)
```

```python
#Berekening om glucose % om te zetten naar glucose mmol/l
medtronic2["BG Reading [mmol/l]"] = 10.93 * ["BG Reading [mmol/l]"] - 23.5
medtronic2
```
