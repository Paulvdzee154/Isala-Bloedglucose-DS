```python
# pandas inporteren, csv file ophalen en checken of dit goed gaat
import pandas as pd
abbott= pd.read_csv("Abbott_US.csv", delimiter=",")
abbott.head()
```

```python
#Data set filteren op de gegevens die ik nodig heb en alle NaN values droppen
abbott2 = abbott[['Subject code number', 'Local date [yyyy-mm-dd]', 'Local time [hh:mm]', 'Historic Glucose [mmol/l]']]
abbott2.dropna(subset = ["Historic Glucose [mmol/l]"], inplace=True)      
abbott2
```

```python
#Data type van Historic Glucose veranderen naar interger en checken of dit goed gegaan in
abbott2 = abbott2.astype({"Historic Glucose [mmol/l]": int})
print(abbott2.dtypes)
```

```python
#Berekening om glucose % om te zetten naar glucose mmol/l
abbott2["Historic Glucose [mmol/l]"] = 10.93 * ["Historic Glucose [mmol/l]"] - 23.5
abbott2
```
