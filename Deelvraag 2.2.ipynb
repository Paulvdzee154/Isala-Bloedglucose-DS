# Data Preparation

```python
#Dataset in tween splitsen (had achteraf niet gehoeven)
df_cv1 = df.copy().drop(columns=['n-2', 'Mean-2', 'Bias-2', 'CV%-2', 'Year'])
df_cv2 = df.copy().drop(columns=['n-1', 'Mean-1', 'Bias-1', 'CV%-1', 'Year'])
```

```python
#CV = (Std/Main) * 100, terugrekenen om de std te krijgen is dus (cv / 100)* mean
df_cv1["Std-1"] = (df_cv1["CV%-1"]/100)
df_cv2["Std-2"] = (df_cv2["CV%-2"]/100)
```

```python
df_cv1["Std-1"] = df_cv1["Std-1"] * df_cv1["Mean-1"]
df_cv2["Std-2"] = df_cv2["Std-2"] * df_cv2["Mean-2"]
```

```python
#Afronden op 1 decimaal zoals de rest van de tabel
df_cv1["Std-1"] = df_cv1["Std-1"].round(1)
df_cv2["Std-2"] = df_cv2["Std-2"].round(1)
```

# Toelichting
Uit de grafiek hierboven kan geconcludeerd worden dat Abott Alinity de meest betrouwbare methode is om te gebruiken, omdat deze zowel de laagste standaarddeviaties als de laagste variatiecofficient heeft in vergelijking met de andere methodes. Een lage standaarddeviatie/CV betekent weinig variatie, wat zorgt voor een hogere consistentie en voorspelbaarheid.
