```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from functools import reduce

%config IPCompleter.greedy=True
```

```python
df_main = pd.read_csv('population.csv', delimiter=';')
df_main = df_main[df_main['Subject code number'] == 903]
```

```python
df_fitbit_sleep = pd.read_csv('Fitbit-sleep.csv')
df_fitbit_sleep.head()
```

```python
df_fitbit_steps = pd.read_csv('Fitbit-steps.csv')
df_fitbit_steps.head()
```

```python
df_main = df_main.merge(df_fitbit_sleep,on='Subject code number')
```

```python
df_main = df_main.merge(df_fitbit_steps,on='Subject code number')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
```

```python
df = df_main
dfkmeans = df.iloc[:,0:len(df.columns)-1]
print(dfkmeans)
index = range(1,10)
inert = []

for clust in index:
    model = KMeans(n_clusters=clust)
    res = model.fit(dfkmeans)
    inert.append(res.inertia_)

sns.lineplot(x=index, y=inert)
```
